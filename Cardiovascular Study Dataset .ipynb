```python
df_train=pd.read_csv("train.csv")
```

```python
df_train.head()
```

```python
#No.of Males and Females
df_train.sex.value_counts()
```

```python
#No.of People smoking out of 3390
df_train.is_smoking.value_counts()
```

```python
#checking for missing values
df_train.isnull().sum()
```

```python
#Removing non significant var
df_train=df_train.drop(['id','education'],axis=1)
```

```python
df_train.head(8)
```

```python
#mode for cat var
df_train.groupby(['BPMeds']).size()

#max value: 0=>3246
```

```python
#filling with mode for Cat Var
df_train['BPMeds']=df_train['BPMeds'].fillna(0.0)
```

```python
#imputing the missing values manually with mean

df_train['cigsPerDay']=df_train['cigsPerDay'].fillna(df_train['cigsPerDay'].mean())
df_train['BMI']=df_train['BMI'].fillna(df_train['BMI'].mean())
df_train['heartRate']=df_train['heartRate'].fillna(df_train['heartRate'].mean())
df_train['glucose']=df_train['glucose'].fillna(df_train['glucose'].mean())
df_train['totChol']=df_train['totChol'].fillna(df_train['totChol'].mean())
```

```python
#checking for missing values
df_train.isnull().sum()
```

```python
df_train.info()
```

```python
#Encoding 
df_train_dummied=pd.get_dummies(df_train,columns=['sex'])
```

```python
df_train_dummied
```

```python
df_train_dummied=pd.get_dummies(df_train_dummied,columns=['is_smoking'])
```

```python
df_train_dummied.head()
```

```python
#removing first dummy
used_features=['age', 'cigsPerDay', 'BPMeds', 'prevalentStroke', 'prevalentHyp',
               'diabetes', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate','sex_M', 'is_smoking_YES']
```

```python
#Indep Variable and Target Variable
x=df_train_dummied[used_features].values
pd.DataFrame(x)
```

```python
y=df_train_dummied['TenYearCHD'].values
pd.DataFrame(y)
```

```python
#train and test data split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=150)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Model Building
logi_model=LogisticRegression()
logi_model.fit(x_train,y_train)
```

```python
#predicted
y_pred=logi_model.predict(x_test)
```

```python
#confusion Matrix
metrics.confusion_matrix(y_test,y_pred)
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
#ACCURACY=85%
```

```python
#cross validation
from sklearn.model_selection import cross_val_score
accuracy=cross_val_score(estimator=logi_model, X=x_train, y= y_train, cv=15)
```
